```python
import matplotlib.pyplot as plt
```

```python
# Syntax for Stack Plot -
# x=[]
# y=[]
# plt.stackplot(x,y)
# plt.show()
# Stack plot is also known as area plot
```

```python
x = [1,2,4,6,9,12]
area1 = [1,2,4,9,11,12]
area2 = [2,3,4,5,6,7]
area3 = [4,5,6,7,8,9]

l = ["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=l,colors=['m','y','g'])

plt.title("Stack Plot",fontsize=16)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# plt.grid()
plt.legend()
plt.show()
```
