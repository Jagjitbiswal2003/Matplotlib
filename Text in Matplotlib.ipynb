```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,1,5,6]

plt.plot(x,y,color='m')

plt.title("Text in Matplotlib",fontsize=17)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.text(1,6,"Hello",fontsize=15,style='italic',bbox={"facecolor":"g"})
plt.annotate('Python',xy=(1,1),xytext=(3,3),arrowprops=dict(facecolor="black",shrink=100))

plt.axis([0,5,0,7])
plt.show()
```
