```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.title("Step Plot",fontsize=19)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.step(x,y,color='black',marker="o",ms=10,mfc='m',label="React")

plt.legend(loc=2)
plt.grid()
plt.show()
```
