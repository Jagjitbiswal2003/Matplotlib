```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,widths=0.1,tick_labels=["Python"],patch_artist=False,showmeans=True,sym=" ",boxprops=dict(color="m"),
            capprops=dict(color="r"),whiskerprops=dict(color="g"))
# notch=True - it change the shape of the box plot
# vert=False - it change the box plot into horizontal direction
# whis=value - it connect the line with outliers
# sym="colorsign" ex-"g+" - it change the color of outliers and the sign change the shape of outliers
plt.title("Box Plot and Whisker Plot")
plt.show()
```

```python
# Multiple box plot-
x = [10,20,30,40,50,60,70]
x1 = [10,20,40,50,30,60,90]
y = (x,x1)
plt.boxplot(y,widths=0.1,tick_labels=["Python","Java"],patch_artist=False,showmeans=True,sym=" ",boxprops=dict(color="m"),
            capprops=dict(color="r"),whiskerprops=dict(color="g"))
plt.title("Multiple Box Plot")
plt.show()
```
