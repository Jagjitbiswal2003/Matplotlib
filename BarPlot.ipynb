```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Syntax for Bar plot -
# x = [] x-parameter
# y = [] y-parameter
# plt.bar(x,y)
#plt.show()
```

```python
x = ["Python","c++","c","Java","React"]
y = [89,80,70,85,90] # popularity of languages out of 100
z=[10,20,30,40,50]
width=0.2
p=np.arange(len(x))
p1 =[j+width for j in p]


plt.xlabel("Languages",fontsize=10) # x-axis name
plt.ylabel("Popularity",fontsize=10) #y-axis name
plt.title("Languages & Popularity",fontsize=20) #plot title


c=['b','y','g','m','gray']
c1=['y','b','m','g','black']
plt.bar(p,y,width,color=c,align="edge",edgecolor="blue",linewidth=2,alpha=0.8,label="Popularity")
plt.bar(p1,z,width,color=c1,align="edge",edgecolor="blue",linewidth=2,alpha=0.8,label="Popularity1") #multiple bar in same frame
plt.xticks(p+width,x,rotation=10)
plt.legend()
# align by default value is center align has also a value i.e edge
plt.show()
```

```python
# Horizontal bar plot

x = ["Python","c++","c","Java","React"]
y = [89,80,70,85,90] # popularity of languages out of 100
z=[10,20,30,40,50]




plt.xlabel("Languages",fontsize=10) # x-axis name
plt.ylabel("Popularity",fontsize=10) #y-axis name
plt.title("Languages & Popularity",fontsize=20) #plot title


c=['b','y','g','m','gray']
c1=['y','b','m','g','black']

plt.barh(x,y,height=0.4,color=c,label="Popularity")
plt.barh(x,z,height=0.4,color=c1,label="Popularity1") #multiple bar in same frame
plt.legend()
plt.show()
```
