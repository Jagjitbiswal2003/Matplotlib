```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="black",bottom=1,basefmt="m",label="Python",)
# orientation="horizontal" - it makes the graph horizontally
# linefmt=":" it makes a dotted line
plt.legend()
plt.show()
```
