```python
x=np.array([0,1,2,3,4,5])
y=np.array([0,1,2,3,4,5])

plt.plot(x,y,color='m')
plt.fill_between(x,y,color='m',where=(x>=2) & (x<=4),alpha=0.4)
#plt.fill_between(x=[2,4],y1=2,y2=4)

plt.title("Fill Between Plot",fontsize=19)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
