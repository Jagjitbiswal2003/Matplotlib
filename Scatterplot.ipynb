```python
x = [1,2,3,4,5,6]
y = [2,4,1,3,5,6]
plt.title("Scatter Plot",fontsize=14)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Number",fontsize=12)
plt.scatter(x,y,color="black")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,4,1,3,5,6]
#color = ['g','r','m','black','gray','pink'] or
colors = [10,49,30,29,56,20]
size = [40,30,30,40,20,33]
plt.title("Scatter Plot",fontsize=14)
plt.xlabel("Day",fontsize=12)
plt.ylabel("Number",fontsize=12)
plt.scatter(x,y,c=colors,s=size,alpha=0.7,cmap='viridis') # alpha value lies in between 0 to 1
t=plt.colorbar()
t.set_label("Color Bar")
plt.show()
```
